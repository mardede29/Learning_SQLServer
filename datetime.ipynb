```sql
select release_date,
year(release_date),
DATEPART(year,release_date) YEAR,
datepart(month,release_date) month,
datepart(day, release_date) day,
DATENAME(WEEKDAY,release_date) weekday,
datename(month,release_date) day,
format(release_date,'d','th'),
format(release_date,'dd','th'),
format(release_date,'ddd','th'),
format(release_date,'dddd','th'),
format(release_date,'M','th'),
format(release_date,'MM','th'),
format(release_date,'MM','th'),
format(release_date,'MMMM','th')
from movie_gross
```

```sql
select format(release_date,'ddd','en') weekday,count(*) [total_day]
 from movie_gross
 group by format(release_date,'ddd','en')
 order by count(*) desc
```
